```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('CRISPR_gene_effect_lung.csv')
df2=pd.read_csv('CRISPR_gene_effect_bile_duct.csv')
df3=pd.read_csv('CRISPR_gene_effect_blood.csv')
df4=pd.read_csv('CRISPR_gene_effect_breast.csv')
df5=pd.read_csv('CRISPR_gene_effect_central_nervous_system.csv')
df6=pd.read_csv('CRISPR_gene_effect_colorectal.csv')
```

```python
df_list = [df1, df2, df3, df4, df5, df6]
```

```python
for df in df_list: 
    cols = [col for col in df.columns if '(' in col]
    df['mediancancer'] = df[cols].median(axis=1)
```

```python
df1.loc['mediangene_Lung'] = df1.median(numeric_only=True)
df2.loc['mediangene_BileDuct'] = df2.median(numeric_only=True)
df3.loc['mediangene_Blood'] = df3.median(numeric_only=True)
df4.loc['mediangene_Breast'] = df4.median(numeric_only=True)
df5.loc['mediangene_CentralNerve'] = df5.median(numeric_only=True)
df6.loc['mediangene_Colorectal'] = df6.median(numeric_only=True)
```

```python
df1=df1.drop(columns='Unnamed: 0')
df2=df2.drop(columns='Unnamed: 0')
df3=df3.drop(columns='Unnamed: 0')
df4=df4.drop(columns='Unnamed: 0')
df5=df5.drop(columns='Unnamed: 0')
df6=df6.drop(columns='Unnamed: 0')
```
