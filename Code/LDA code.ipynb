```python
display(y.value_counts())
```

```python
X1 = X["essentiality"]
X2 = X["mutation"]
X3 = X['expression']
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda = QuadraticDiscriminantAnalysis()
```

```python
from matplotlib import pyplot as plt
```

```python
y_bladder_n = y == 'negative'
y_bladder_p = y == 'positive'
```

```python
X_train = X[["essentiality","mutation",'expression','x_25','x_10','x_19','x_5','x_1','x_3','x_2','x_22','x_31','x_6','x_21','x_30']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
lda = LDA(n_components=1)

X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

```python
clf = LinearDiscriminantAnalysis()
 
#checking for the model accuracy using score method
clf.fit(X_train, y_train).score(X_train, y_train)
```

```python
y_test = y_test.map({'positive': 1, 'negative': 0}).astype(int)
```

```python
y_pred_proba = qda.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
```

```python
auc = metrics.roc_auc_score(y_test, y_pred_proba)
```

```python
plt.plot(fpr,tpr,label='bladder'+"AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend()
plt.show()
```

```python

 
#plotting for roc curve
plt.plot(fpr, tpr, color="navy", linestyle="--", label = "roc_curve = %0.2f"% auc)
plt.legend(loc = "upper center")
 
```
