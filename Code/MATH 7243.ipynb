```python
training_csv_files = []
for file_ in os.scandir():
    if file_.name.endswith('csv'):
        training_csv_files.append(file_.name)
```

```python
all_genes_ = df.index
```

```python
for csv_file in training_csv_files:
    train_df = pd.read_csv(csv_file)
    genes_ = train_df.iloc[:,0]
    ava_genes_ = [ x for x in genes_ if x in all_genes_ ]
    df.loc[ava_genes_, :].to_csv(csv_file.split('_')[0] + '_gene_expressions.csv')
```
